# Dummy Data

```python
def get_dummies(data):
    data_dummies = pd.get_dummies(data,drop_first=True)
    return data_dummies
```

```python
data = raw_data.copy()
```

```python
data
```

```python
gender_dummies = get_dummies(data.gender)
```

```python
gender_dummies
```

```python
data_with_gender_dummies = pd.concat([data,gender_dummies],axis=1)
```

```python
data_with_gender_dummies
```

```python
data_drop_gender = data_with_gender_dummies.drop(labels=['gender'],axis=1)
```

```python
data_drop_gender
```

```python
senior_citizen = get_dummies(data_drop_gender.SeniorCitizen)
```

```python
senior_citizen.columns = ['isCitizen']
```

```python
senior_citizen
```

```python
data_with_senior_dummies = pd.concat([data_drop_gender,senior_citizen],axis=1)
```

```python
data_drop_senior_citizen = data_with_senior_dummies.drop(labels=['SeniorCitizen'],axis=1)
```

```python
data_drop_senior_citizen
```

```python
partner_dummy = get_dummies(data_drop_senior_citizen.Partner)
partner_dummy.columns = ['HavePartner']
```

```python
partner_dummy
```

```python
data_with_senior_dummies = pd.concat([data_drop_senior_citizen,partner_dummy],axis=1)
data_drop_partner = data_with_senior_dummies.drop(labels=['Partner'],axis=1)
data_drop_partner
```

```python
dependents_dummy = get_dummies(data_drop_partner.Dependents)
dependents_dummy.columns = ['HaveDepedents']
```

```python
data_with_dependents_dummies = pd.concat([data_drop_partner,dependents_dummy],axis=1)
data_drop_dependents = data_with_dependents_dummies.drop(labels=['Dependents'],axis=1)
data_drop_dependents
```

```python
phone_services_dummy = get_dummies(data_drop_partner.PhoneService)
phone_services_dummy.columns = ['PhoneServices']
```

```python
data_with_phone_services_dummies = pd.concat([data_drop_dependents,phone_services_dummy],axis=1)
data_drop_phones = data_with_phone_services_dummies.drop(labels=['PhoneService'],axis=1)
data_drop_phones
```

```python
MultipleLines = get_dummies(data_drop_partner.MultipleLines)
MultipleLines
```

```python
MultipleLines.columns = ['MultiplineLines_No_phone_service','Multiplines_YES']
```

```python
data_with_multiplelines_dummies = pd.concat([data_drop_phones,MultipleLines],axis=1)
data_drop_multiplelines = data_with_multiplelines_dummies.drop(labels=['MultipleLines'],axis=1)
data_drop_multiplelines
```

```python
InternetService = get_dummies(data_drop_partner.InternetService)
InternetService
```

```python
InternetService.columns = ['InternetService_Fiber_Optic','InternetService_No']
data_with_internet_services_dummy = pd.concat([data_drop_multiplelines,InternetService],axis=1)
data_drop_internet_services = data_with_internet_services_dummy.drop(labels=['InternetService'],axis=1)
data_drop_internet_services
```

```python
OnlineSecurity = get_dummies(data_drop_partner.OnlineSecurity)
OnlineSecurity
```

```python
OnlineSecurity.columns = ['OnlineSecurity_No_Internet_Service','OnlineSecurity_Yes']
data_with_onlines_security_dummy = pd.concat([data_drop_internet_services,OnlineSecurity],axis=1)
data_drop_onlines_security = data_with_onlines_security_dummy.drop(labels=['OnlineSecurity'],axis=1)
data_drop_onlines_security
```

```python
OnlineBackup = get_dummies(data_drop_partner.OnlineBackup)
OnlineBackup
```

```python
OnlineBackup.columns = ['OnlineBackup_No_Internet_Service','OnlineBackup_Yes']
data_with_onlines_backup_dummy = pd.concat([data_drop_onlines_security,OnlineSecurity],axis=1)
data_drop_onlines_backup = data_with_onlines_backup_dummy.drop(labels=['OnlineBackup'],axis=1)
data_drop_onlines_backup
```

```python
DeviceProtection = get_dummies(data_drop_partner.DeviceProtection)
TechSupport = get_dummies(data_drop_partner.TechSupport)
StreamingTV = get_dummies(data_drop_partner.StreamingTV)
StreamingMovies = get_dummies(data_drop_partner.StreamingMovies)
Contract = get_dummies(data_drop_partner.Contract)
PaperlessBilling = get_dummies(data_drop_partner.PaperlessBilling)
PaymentMethod = get_dummies(data_drop_partner.PaymentMethod)
```

```python
DeviceProtection
```

```python
DeviceProtection.columns = ['DeviceProtection_No_Internet_Service','DeviceProtection_Yes']
```

```python
TechSupport
```

```python
TechSupport.columns = ['TechSupport_No_Internet_Service','TechSupport_Yes']
```

```python
StreamingTV
```

```python
StreamingMovies
```

```python
PaperlessBilling
```

```python
PaymentMethod.columns = ['PaymentMethod_Credit_card','PaymentMethod_Electronic_Check','PaymentMethod_Mailed_Check']
```

```python
data_with_dummy = pd.concat([data_drop_onlines_backup,DeviceProtection,TechSupport,StreamingTV,Contract,PaperlessBilling,
                             PaymentMethod],axis=1)
```

```python
data_with_dummy
```

```python
data_done_with_dummy_data = data_with_dummy.drop(labels=['DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
                                                          'Contract', 'PaperlessBilling', 'PaymentMethod'],axis=1)
```

```python
data_done_with_dummy_data
```
