```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.set_option('display.max_columns', None)
```

```python
raw_data = pd.read_csv('dataset.csv')
```

```python
raw_data.TotalCharges = raw_data.TotalCharges.replace(' ',np.nan)
raw_data.TotalCharges = raw_data.TotalCharges.astype('float')
```

```python
data = raw_data[raw_data.TotalCharges.notnull()]
```

```python
data
```

```python
data.Churn.value_counts()
```

```python
data = data.reset_index()
```

```python
data
```

```python
zero_targets_counter = 0
indices_to_remove = []

for i in range (data.shape[0]):
  if data.iloc[i,:].Churn=='No':
    zero_targets_counter += 1
    if zero_targets_counter >1869:
      indices_to_remove.append(i)
```

```python
data = data.drop(labels=indices_to_remove)
```

```python
data = data.drop(labels=['index','customerID','MonthlyCharges','TotalCharges'],axis=1)
```

```python
data.to_csv('preprocessing3.csv',index=False)
```
